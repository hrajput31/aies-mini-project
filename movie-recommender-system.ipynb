```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
# Columns to be kept
#Genres #ID #Keywords #Titles #Overview #Cast #Crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def con3(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(con3)
```

```python
movies.head()
```

```python
def fetch_dir(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
df=movies[['movie_id','title','tags']]
df
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(txt):
    y=[]
    for i in txt.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
v=cv.fit_transform(df['tags']).toarray()
```

```python
v
```

```python
cv
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(v)
```

```python
similarity
```

```python
df["title"]=='Avatar'
```

```python
def recommend(mov):
    mov_ind=df[df['title']==mov].index[0]
    d=similarity[mov_ind]
    mov_list=sorted(list(enumerate(d)),reverse=True,key=lambda x:x[1])[1:6]
    for i in mov_list:
        print(df.iloc[i[0]].title)
```

```python
recommend("The Master of Disguise")
```
